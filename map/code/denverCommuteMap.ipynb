```python
# Define function to allow us to display charts in an iframe


def show_iframe(url, iheight=400, iwidth=1000):
    display_string = '<iframe src={url} width={w} height={h}></iframe>'.format(url=url, w=iwidth, h=iheight)
    print(display_string)
    return ipd.HTML(display_string)
```

```python
# Read in chicago zips into a GeoDataFrame
den_df = gpd.read_file('../data/cb_2016_us_zcta510_500k/denver_zips_2016.geojson').rename(columns = {'ZCTA5CE10': 'zip_code'})

# Set the projection to be consistent with Bokeh tiles (Mercator)
den_df = den_df.to_crs({'init': 'epsg:3857'})

# Set zip code to index to facilitate joining
for dataframe in [df, den_df]:
    if 'zip_code' in dataframe.columns:
        # Make sure zip codes are integers in both sets
        dataframe['zip_code'] = dataframe['zip_code'].astype(int)
        dataframe.set_index('zip_code', inplace=True)
        
# Join the access data to the spatial data
for c in df.columns:
    den_df[c] = df[c].loc[den_df.index]
```

```python
# mean_commute replaces overall here
# Capture minimum and maximum values within each driver bucket
overall_min = 3.611249999 # three standard deviations from the mean
overall_max = 48.324108 # three standard deviations from the mean

# Capture normalizer objects (they map data to the [0,1] interval)
norm = {
    'mean_commute': mpl.colors.Normalize(vmin=overall_min, vmax=overall_max)   
}

# Generate color mappers
cmapper = {
    'mean_commute': cmap.ScalarMappable(norm=norm['mean_commute'], cmap='seismic')
}

# Map colors to values
den_df['overall_color'] = den_df['mean_commute'].apply(lambda x: mpl.colors.to_hex(cmapper['mean_commute'].to_rgba(x)))
```

```python
# Define data source
den_commute_src = bm.GeoJSONDataSource(geojson=den_df.reset_index().to_json())

def plot_dist(v, height_in=900, width_in=750):
    # Define location of output file
    outf = '../figs/denver_{}.html'.format(v)
    bp.output_file(outf)
    # Define figure
    fig = bp.figure(title='Denver {}'.format(v.title()),
                       height=height_in, width=width_in)
    # Add zip polygons
    fig.patches(xs='xs', ys='ys', alpha=.8, source=den_commute_src,
                fill_color={'field': 'overall_color'}, line_width=1, line_color='#626770')
    # Define hover appearance (in raw HTML)
    hover_html = '''
    <div><b>ZIP Code</b>: @zip_code</div>
    <div><b>Mean Commute Time</b>: @{}</div>
    '''.format(v)
    hover = bm.HoverTool(
        point_policy='follow_mouse',
        tooltips=(hover_html)
    )
    # Define tap tool
    tap = bm.TapTool()
    fig.add_tools(hover, tap)
    # Add map tile
    fig.add_tile(CARTODBPOSITRON_RETINA)
    # Get rid of axis lines
    fig.xaxis.visible = False
    fig.yaxis.visible = False
    fig.grid.visible = False
    # Show the figure
    bp.save(fig)
    return outf

ovr = plot_dist('mean_commute')
show_iframe(ovr, iwidth=850, iheight=950)
```
